## Q 30-01

Este es el problema del intercambio de monedas.

Tienes una moneda de 400 wones.

Escriba el resultado de cómo el algoritmo `coin_change()` determina el cambio de la moneda de 800 wones.
```
coins  = [500, 400, 100, 50, 10]
amount = int(input("Ingrese la cantidad: "))
changes= coin_change(coins, amount)
print(changes, len(changes))

Ingrese la cantidad: 800
```

````python
# Ejemplo de cambio de monedas

def coin_change(coins, amount):
    changes = []
    largest = 0
    
    while amount > 0:
        if coins[largest] <= amount:         # Verifica si la primera moneda cabe en amount
            changes.append(coins[largest])  # Agrega la moneda a la lista de cambios
            amount -= coins[largest]        # Resta el valor de la moneda a amount
        else:
            largest += 1                    # Si la moneda no cabe, pasa a la siguiente
    return changes




# establezco la lista de elementos que componen la solución

coins = [500, 400, 100, 50, 10]              # monedas disponibles
amount = int(input("Ingrese la cantidad de dinero que desea cambiar: "))
print(f"El cambio de {amount} es {coin_change(coins, amount)} con {len(coin_change(coins, amount))} monedas.")



# En el caso de este ejemplo, vemos que con 500, 400, 100, 50, 10, el cambio de 750 es [500, 100, 100, 50] con 4 monedas.
# El algoritmo determina el cambio de la moneda realizando una resta de la cantidad de dinero con el valor de la moneda más grande,
# y si el resultado es mayor a 0, agrega la moneda a la lista de cambios y resta el valor de la moneda a la cantidad de dinero.
````

## Q 31-01

Tienes 8 monedas idénticas numeradas del 1 al 8. De estas, solo una moneda es más pesada que la otra.

Suponiendo que puede pesar monedas con una balanza equilibrada, diseñe un algoritmo que encuentre la moneda más pesada.

En este momento, ¿cuántas veces se debe usar al menos la balanza?

Para realizar este ejemplo podríamos utilizar el método de divide y conquista, para ello realizamos lo siguiente:
    1. Dividimos en dos grupos de cuatro monedas
    2. Pesamos los dos grupos, lo que equivale a un uso en total.
    3. Tomamos el grupo más pesado y lo dividimos en dos grupos de dos monedas. sabemos que la más pesada debe estar allí porque las demás son identicas.
    4. pesamos los dos grupos y elegimos el más pesado, esto equivale a dos usos en total.
    5. pesamos las dos monedas restantes y encontramos nuestra moneda, lo que equivale a 3 usos en total.

    Necesitariamos tres usos de la balanza para encontrar la moneda.

````python
def fib4(n):
    if n <= 1:
        return n
    else:
        a,b = 0,1
        for i in range(2, n + 1):
            a,b = b,a+b
        return b
    
def fib3(F, n):
    if n <= 1:
        return F[n]
    else:
        if F[n] < 0:
            F[n] = fib3(F, n - 1) + fib3(F, n - 2)
        return F[n]

F = {0: 0, 1: 1}

def fib2(n):
    if n in F:
        return F[n]
    else:
        F[n] = fib2(n - 1) + fib2(n - 2) 
        return F[n]
    
def fib1(n):
    if n == 0 or n == 1:
        return n
    else:
        return fib1(n - 1) + fib1(n - 2)    


N = int(input("Ingrese un numero: "))
print(fib4(N))
````

`fib4()` es la más eficiente porque utiliza un enfoque iterativo que minimiza el uso de memoria y reduce el tiempo de ejecución al evitar llamadas recursivas y recalculaciones. Las funciones `fib2()` y `fib3()` son eficientes gracias a la memorización que guarda resultados previos, pero usan más memoria por almacenar estos resultados. `fib1()` es ineficiente para valores grandes porque realiza muchas recalculaciones innecesarias debido a su enfoque recursivo sin memorización.

````python
# Analice el resultado de la ejecución del siguiente código 'bin3()' y compárelo con el rendimiento de las funciones `bin1()` y `bin2(

def binl(n, k):
    if k == 0 or n == k:
        return 1
    else:
        return binl(n - 1, k - 1) + binl(n - 1, k)
    
def bin2(n, k):
    B= [[0] * (i + 1) for  i in range(n + 1)]
    for i in range(n + 1):
        for j in range(i + 1):
            if j == 0 or j == i:
                B[i][j] = 1
            else:
                B[i][j] = B[i - 1][j - 1] + B[i - 1][j]
    return B[n][k]

def bin3(n,k):
        B = [0] * (n + 1)
        for i in range(n + 1):
            for j in range(n, -1, -1):
                if j==0 or j==i:
                    B[j] = 1
                else:
                    B[j] = B[j] + B[j-1]
        return B[k]

for i in range(10):
        for j in range(i+1):
            print(bin3(i,j),end=" ")
        print()
````

___

`bin3()` es la más eficiente en términos de uso de memoria porque utiliza un arreglo unidimensional para calcular los coeficientes binomiales, actualizándolo en cada paso, en lugar de almacenar toda la tabla de valores como `bin2()`. Aunque `bin1()` y `bin2()` pueden ser efectivos para casos pequeños, `bin1()` sufre de una eficiencia significativamente baja para valores grandes debido a su enfoque recursivo que realiza muchas recalculaciones. `bin2()` mejora en tiempo al evitar recalculaciones mediante programación dinámica, pero `bin3()` ofrece la mejor optimización de memoria sin comprometer el tiempo de ejecución.

````python
def is_valid_sum_puzzle(grid, row, col, num, target_sum=12):
    
    row_sum = 0
    col_sum = 0
    
    for x in grid[row]:  # realizar la suma de la fila
        if x is not None: # si el espacio esta vacio no lo sume
            row_sum += x
    
    for i in range(len(grid)): # realizar la suma de la columna
        if grid[i][col] is not None: # si el espacio esta vacio no lo sume
            col_sum += grid[i][col]
        
  
    # Verificar la suma de la fila y la columna con el número añadido
    if (row_sum + num > target_sum) or (col_sum + num > target_sum):
        return False
    
    # Verificar si el número ya está en la fila o columna
    if num in grid[row] or num in [grid[i][col] for i in range(len(grid))]:
        return False
    
    return True



def solve_puzzle(grid, available_numbers, target_sum=12):
    if not available_numbers:  # Si no hay números disponibles, se ha resuelto el puzzle
        return True

    for row in range(len(grid)):
        for col in range(len(grid[row])):
            if grid[row][col] is None:  # Espacio vacío
                for num in available_numbers.copy():  # Hacer una copia para iterar sobre ella
                    if is_valid_sum_puzzle(grid, row, col, num, target_sum):
                        grid[row][col] = num
                        new_available_numbers = available_numbers.copy()
                        new_available_numbers.remove(num)  # Retirar el número usado de la lista
                        if solve_puzzle(grid, new_available_numbers, target_sum):
                            return True
                        grid[row][col] = None  # Backtrack
                return False  # No se encontró un número válido, por lo tanto se hace backtrack
    return True  # Se llenó el tablero correctamente

# Asumiendo que el 0 ya está colocado y no debe ser usado de nuevo
available_numbers = [3, 0, 2, 4, 7]

# Iniciar el tablero con `None` en los espacios vacíos
grid = [
    [None, None, None],
    [None, None, 8],
    [5, 6, 1]
]

# Llamar a la función de resolución
if solve_puzzle(grid, available_numbers):
    for row in grid:
        print(row)
else:
    print("No solution found")
````
